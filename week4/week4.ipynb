```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# physical paramaters of cooling
# in cgs
p = {
    "a" : 10.0,
    "b" : -0.5,
    "T0" : 20_000.0,
    "Tinit" : 1e7,
    "lambda0" : 1e-22,
    "nH" : 1.0,
}

sim = {
    "dt" : 1e10,
    "t" : 0.0,
    "T" : 0.0,
    "tmin": 6000.0,
    "NMAX" : 10**6,
    "dTmax" : 50,
}

kB = 1.38e-16
```

```python
def L(T):
    L = 0;
    if (T <= p["T0"]):
        #print("cold")
        L = p["lambda0"] * np.power(T/p["T0"], p["a"])
    else:
        #print("hot")
        L = p["lambda0"] * np.power(T/p["T0"], p["b"])
    #print(L)
    return L

def cooling(T):
    dTdt = -2.0/(3.0 * kB) * p["nH"] * L(T)
    return dTdt
```

```python
def RK2(y, h, eq):
    k1 = eq(y)
    #print(type(k1))
    k2 = eq(y + 0.5*h*k1)
    return y + 0.5 * k2 * h
```

```python
def simulate(integ, dt):
    sim["t"] = 0
    N = 0
    sim["T"] = p["Tinit"]
    Ts, ts = [], []
    while (sim["T"] > sim["tmin"]) and (N < sim["NMAX"]):
        #print(sim["T"])
        #print(sim["dt"])
        sim["T"] = integ(sim["T"], dt, cooling)
        Ts.append(sim["T"])
        ts.append(sim["t"])
        sim["t"] += dt
        N += 1
    return ts, Ts
```

```python
ts_n, Ts_n = simulate(RK2, sim["dt"])
plt.plot(ts_n, Ts_n)
plt.yscale("log")
plt.ylabel("T (K)")
plt.xlabel("time (s)")
print(len(ts_n))

# (b) We need roughly 100 000 steps to reach the final temperature
```

```python
dts = np.logspace(10.6, 10.8, 5)
for dt in dts:
    ts, Ts = simulate(RK2, dt)
    plt.plot(ts, Ts, label=f"{dt:.3e}")
plt.yscale("log")
plt.legend()

# (b) The calculation starts becoming unstable after dt > 4 * 10**10 s
```

```python
def simulate_adaptive(integ, dt):
    
    eps = 0.1
    
    sim["t"] = 0
    N = 1
    sim["T"] = p["Tinit"]
    ts, Ts, dts = [0], [p["Tinit"]], [1e10]
    while (sim["T"] > sim["tmin"]) and (N < sim["NMAX"]):
        Ta = integ(sim["T"], dt/2, cooling)
        Tb = integ(Ta, dt/2, cooling)
        
        Tc = integ(sim["T"], dt, cooling)
        
        Terr = abs(Tb - Tc)
        if Terr > sim["dTmax"] : 
            dt *= (1 - eps)
        else:
            dt *= (1 + eps)
        
        N += 1
        sim["t"] += dt
        sim["T"] = Tc
        Ts.append(Tc)
        dts.append(dt)
        ts.append(sim["t"])
    
    return ts, Ts, dts
```

```python
ts_a, Ts_a, dts_a = simulate_adaptive(RK2, 1e10)
ts_b, Ts_b, dts_b = simulate_adaptive(RK2, 1e11)
ts_c, Ts_c, dts_c = simulate_adaptive(RK2, 1e12)
ts_d, Ts_d, dts_d = simulate_adaptive(RK2, 1e13)

plt.plot(ts_a, Ts_a, label="adaptive, dt = 1e10")
plt.plot(ts_b, Ts_b, label="adaptive, dt = 1e11")
plt.plot(ts_c, Ts_c, label="adaptive, dt = 1e12")
plt.plot(ts_d, Ts_d, label="adaptive, dt = 1e13")

plt.plot(ts_n, Ts_n, label="non adaptive")
plt.legend()
plt.yscale("log")
plt.ylabel("T (K)")
plt.xlabel("time (s)")
print(len(ts_a))

# robust upto more than 1000 times larger timesteps. and can be be tuned using the epsilon parameter
# only around 500 steps are needed w. dt = 1e10 s, depending on epsilon parameter
```
